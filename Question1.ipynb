```python
import pandas as pd

# Load the dataset from your gist link
archana05= pd.read_csv("https://gist.githubusercontent.com/archana340/8002f4417c0e4f66519c08de563471ce/raw/8ee8ae4703a8fafe3799daacbdf9fc927e300982/data1.csv")

# Display the last few rows of the dataframe
print(archana05.tail())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Set style
sns.set(style="whitegrid")

# Plot Population
plt.figure(figsize=(10, 5))
sns.barplot(x='Year', y='Population', data=archana05, color='skyblue')
plt.title("Frequency Distribution: India's Population")
plt.ylabel("Population")
plt.xlabel("Year")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot GDP
plt.figure(figsize=(10, 5))
sns.barplot(x='Year', y='GDP_Trillion_USD', data=archana05, color='lightgreen')
plt.title("Frequency Distribution: India's GDP (Trillion USD)")
plt.ylabel("GDP (Trillion USD)")
plt.xlabel("Year")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot Literacy Rate
plt.figure(figsize=(10, 5))
sns.barplot(x='Year', y='Literacy_Rate', data=archana05, color='salmon')
plt.title("Frequency Distribution: India's Literacy Rate (%)")
plt.ylabel("Literacy Rate (%)")
plt.xlabel("Year")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# Show plot
plt.show()
```

```python
# Pairplot to visualize pairwise relationships in the dataset.
sns.pairplot(archana05, hue='Year', height=2.5)
plt.savefig('new.png')
plt.show()
```

```python
import altair as alt
# Compute correlation matrix
# GDP chart
gdp_chart = alt.Chart(archana05).mark_line(point=True).encode(
    x='Year:O',
    y='GDP_Trillion_USD:Q',
    color=alt.value('green')
).properties(title='India GDP Over the Years')

# Literacy chart
literacy_chart = alt.Chart(archana05).mark_line(point=True).encode(
    x='Year:O',
    y='Literacy_Rate:Q',
    color=alt.value('blue')
).properties(title='India Literacy Rate Over the Years')

# Combine vertically
combined = alt.vconcat(gdp_chart, literacy_chart)

# Display
combined.show()
```

```python
#The following packages and libraries form the foundation of this lab’s interactive visualization creation:
#1. Altair
Altair serves as a Python-based statistical visualization library that functions through Vega and Vega-Lite framework. This library provides straightforward usage since it combines well with Jupyter notebooks as well as other Python environments.

Key Features:

The system provides capabilities to generate interactive charts that include line charts and bar charts as well as scatter plots and additional styles.

Drawing field-related data such as x, y coordinates and color, shape and size values assists users in creating effective visualizations.

Users can use hover, click and selection options for interactive features

#2.Matplotlib (Optional for static visuals)
Matplotlib functions as a Python library which users widely apply to generate interactive visualizations alongside static and animated output. The default interface of Altair does not provide interactive features yet developers can implement interactivity by using mpl_toolkits or combining tools with other libraries.

Key Features:

The library allows users to produce multiple graphical options which include histograms and scatterplots and bar charts among others.

It offers excellent static visualization capabilities but users can obtain interactivity through widgets.
```

```python
# Here I am creating:

A line chart for GDP over the years.

A scatter plot for GDP vs Literacy Rate.

A bar chart for Population over the years.
```

```python
# Line chart for GDP over the years
gdp_line_chart = alt.Chart(archana05).mark_line(point=True).encode(
    x='Year:O',
    y='GDP_Trillion_USD:Q',
    color='Year:O',
    tooltip=['Year', 'GDP_Trillion_USD']
).properties(
    title="GDP Over the Years",
    width=500,
    height=300
)

# Scatter plot for GDP vs Literacy Rate
gdp_vs_literacy = alt.Chart(archana05).mark_circle(size=100).encode(
    x='GDP_Trillion_USD:Q',
    y='Literacy_Rate:Q',
    color='Year:O',  # Color based on Year
    tooltip=['Year', 'GDP_Trillion_USD', 'Literacy_Rate']
).properties(
    title="GDP vs Literacy Rate",
    width=500,
    height=300
)

# Bar chart for Population over the years
population_bar_chart = alt.Chart(archana05).mark_bar().encode(
    x='Year:O',
    y='Population:Q',
    color='Year:O',
    tooltip=['Year', 'Population']
).properties(
    title="Population Over the Years",
    width=500,
    height=300
)

# Combine all charts vertically with an interactive legend
combined_charts = alt.vconcat(
    gdp_line_chart,
    gdp_vs_literacy,
    population_bar_chart
).add_selection(
    alt.selection_multi(fields=['Year'], bind='legend')
)

# Display the combined chart with an interactive legend
combined_charts.show()
```
